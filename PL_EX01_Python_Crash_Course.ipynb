## Sequences

- Sequence types (strings, tuples, lists) organize values sequentially, indexes starting from `0`

- Elements are accessed by *indexing*
  - `s[i]` for element `i`

- Sequences can also be *sliced* to obtain a sub-collection
  - `s[i:j]` for elements from `i` up to `j`, exclusive
  - `s[i:j:k]` for elements from `i` up to `j`, exclusive, with increments of `k`
  - indexes may be omitted, `s[:i]` or `s[i:]`

- Negative indexes start from the end of the sequence

- Concatenation with `+` operator
  - Also supports incremental assignments with `+=`

- Comparison operators are defined, use an lexicographic order

- Sequence repetition with `*`

### Tuples

- Tuples are heterogeneous sequences (can contain elements of different types)

- Like strings, tuples are immutable

- Support most sequence operations, except those specific to text manipulation

- Tuples are delimited by parentheses `(` `)`:
  - Empty tuple: `()`
  - Single-element tuple: `(x,)` (comma is required)
  - Can be omitted in certain situations

- Immutability makes tuples suitable as keys in dictionaries or elements of sets

````python
student = ("Frodo Baggins", 2968, "The Shire", True)
print(len(student))           # length
print(True in student)        # membership
print("The Shire" in student) # membership
print(student[0])             # indexing
print(student[0:3])           # slicing
````

````python
# iteration
def final(grades):
  total = 0
  for grade in grades:
    if grade < 10:
      return "R"
    total += grade
  average = total / len(grades)
  return average

print(final((13,15,11))) # one parenthesis for function call, another for tuple definition
print(final((13,1,11)))
````

````python
def add_matrices(m1, m2):
  if len(m1) != len(m2) or len(m1[0]) != len(m2[0]):
    return None
  result = []
  for row in range(len(m1)):
    line = []
    for col in range(len(m1[0])):
      line.append(m1[row][col] + m2[row][col])
    result.append(line)

  return result

add_matrices([[3,5,4], [2,4,4], [2,5,6]],[[1,5,8], [2,0,4], [3,6,0]])
````

## Aliasing

- Aliasing occurs when multiple variables refer to the same value

- Assignment in Python never copies values; it only creates new bindings

- For mutable values (such as lists), a change through one variable affects all others referring to that value

- Aliasing impacts operations such as assignment and (shallow) copying of a list

````python
a = [1,2,3]
b = a
a[1] *= 10
print(b)
b.append(-1)
print(a)
````

````python
students = [
    ["A", 1998, "X"],
    ["B", 1999, "Y"]
]
# make a copy of the outer list
students_copy = students.copy()
students_copy.pop()
print(students)
print(students_copy)

# but for the elements, it just copies the values
students_copy[0][0] = "??"
print(students)
````

## Dictionaries

- Dictionaries (`dict`) are associative collections mapping *keys* to *values*

- Keys must be unique, values may be repeated

- Keys must be hashable (immutable types such as strings, numbers, tuples)

- This makes dictionary access very efficient

- Values can be of any type

- Access, insertion, and update are performed using keys (see [documentation](https://docs.python.org/3/library/stdtypes.html#mapping-types-dict))

- Iteration operations act on keys
  - Method `values()` or `items()` to also iterate values

- Dictionaries are delimited by curly brackets `{` `}`, key-value pairs defined with `:`
  - The empty dictionary is `{}`

- Commonly used to represent mappings, environments, and symbol tables


````python
# dictionary definition
d = { "John": 10, "Jane": 20, "Mary": None }
print(d)
````

````python
# access
print(d["John"])
````

````python
# accessing non-existing keys throws an exceptions
print(d["JOHN"])
````

````python
# to avoid exceptions, use get() with default value
print(d.get("JOHN", "???"))
````

````python
# insertion
d["Mike"] = 5
print(d)
````

````python
# update
d["Mike"] = 15
print(d)
````

````python
# removal
del d["Mike"]
print(d)
````

````python
# removal with value returned
v = d.pop("Jane")
print(v)
````

````python
# iteration
for k in d:
  print(k, "has", d[k], "things")
````

````python
# iteration over key/value pairs
for k, v in d.items():
  print(k, "has", v, "things")
````

## Sets

- Sets (`set`) are unordered collections of unique elements

- Elements must be hashable

- Support standard set operations: union `|`, intersection `&`, difference `-`

- Membership testing is very efficient

- Sets are delimited by curly brackets `{` `}`
  - The empty dictionary is `set()` (because `{}` is a dictionary)

- Many useful operations (see [documentation](https://docs.python.org/3/tutorial/datastructures.html#sets))

- Useful for eliminating duplicates and representing collections of states or symbols


````python
# set definition
seen = { "Movie", "Movie II", "Movie: The Return" }
print(seen)
````

````python
# adding elements
seen.add("Movie Yet Again?")
print(seen)
````

````python
# no repetition
seen.add("Movie II")
print(seen)
````

````python
# cannot select specific element, no order
print(seen[2])
````

````python
# set operations
seen_also = { "Movie", "Enough of Movie", "Movie: The Return" }
print(seen | seen_also)
print(seen & seen_also)
print(seen - seen_also)
````

## ðŸ”¬ Iterators

- An *iterator* is an abstract object that represents a stream of elements from a collection, producing one element at a time

- All Python collections are iterable, meaning they can provide an iterator via `iter(collection)`

- Iterators are stateful: they remember their current position, and only move forward

- Retrieve elements one by one using `next()`, raises `StopIteration` when exhausted

- `for` loops internally use iterators provided by iterables

- The `itertools` module extends this abstraction with composable and lazy iterators,
   - infinite sequences (`count()`, `cycle()`)
   - combinatorial tools (`chain()`, `islice()`, `combinations()`, `permutations()`)

````python
l = [1,2,3]
it = iter(l)
print(it)
````

# Functional programming in Python

- Python supports a functional programming style alongside imperative and object-oriented ones

- Functions are first-class values

- Emphasis on expressions rather than statements, *what* to do rather than *how* to do it

- Functional constructs are often used to process collections and streams of data


## Recursion

- A function can call itself, directly or indirectly

- Recursive functions must define at least one base case to terminate

- Each recursive call creates a new stack frame with its own local variables

- Python does not perform tail-call optimization

- The maximum recursion depth is limited and can be inspected or changed (`sys.getrecursionlimit()`)

````python
def sum_rec(lst):
  if lst == []:
    return 0
  else:
    return lst[0] + sum_rec(lst[1:])

print(sum_rec([5,2,5,7,8]))
````

## Higher-order functions

- Functions can be passed as arguments to other functions

- Functions can be returned as results of other functions

- Sometimes it is useful to use anonymous, `lambda` functions

- Many useful opererations in `functools` module (see [documentation](https://docs.python.org/3/library/functools.html))

- Module `operator` provides operators as functions to be used as values (see [documentation](https://docs.python.org/3/library/operator.html))
  - E.g., operator `+` is function `operator.add()`

- Higher-order functions enable abstraction over behavior

````python
def deriv(f,h):
  """
  Returns function df(x) that approximates the derivative of f(x) for a given
  small h.
  """
  def df(x):
    return round((f(x+h) - f(x))/h, 3)
  return df

# derivative of xÂ² should be (approximately) 2x
def sq(x): return x*x

d_sq = deriv(sq,0.0001)

print("f(10):", d_sq(10))
print("df(10):", d_sq(20))
````

## Sorting lists

- Method `sort()` is a modifier that sorts a list

- It can take, as an optional argument, a `key` function to determine the sorting criteria

- This makes it a higher-order function

- Builti-in function `sorted()` returns a new sorted list, rather than modifying the existing

````python
fruits = [("Kiwis", 40), ("Apples", 200), ("Pears", 150), ("Bananas", 120)]
# use default order on pairs
fruits.sort()
print(fruits)
````

````python
# define second element as key
def second(pair):
  return pair[1]
fruits.sort(key=second)
print(fruits)
````

````python
# define negative second element as key
fruits.sort(key=lambda pair: -pair[1])
print(fruits)
````

````python
# sort() is a modifier
print(fruits.sort())

# sorted() creates a new list
print(sorted(fruits))
````

## Maps, filters, folds

- Built-in higher-order fuctions to manipulate collections

- `map(f, xs)` applies a function to each element of a collection

- `filter(p, xs)` selects elements satisfying a predicate

- These functions are lazy, return iterators that can only be used once

- `reduce(f, xs)` is the fold operation, combines elements using a binary function (needs `functools` module)
  - optional initializer for the result as `reduce(f, xs, i)`, otherwise first element of collection



````python
fruits = [("Kiwis", 40), ("Apples", 200), ("Pears", 150), ("Bananas", 120), ("Oranges", 90)]
double = map(lambda pair: 2*pair[1], fruits)
print(double)
double_list = list(double)
print(double_list)
print(double)
````

````python
lows = filter(lambda pair: pair[1] < 100, fruits)
print(list(lows))
````

````python
double_lows = map(lambda pair:2*pair[1], filter(lambda pair: pair[1] < 100, fruits))
print(list(double_lows))
````

````python
from functools import reduce

def aggregate(rest, new):
  fruits, total = rest
  return fruits + [new[0]], total + new[1]

agg = reduce(aggregate, fruits, ([],0))

print(agg)
````

## Comprehensions

- Compact syntax for constructing collections (e.g., lists or dictionaries)

- Combine mapping and filtering in a single expression

- Often clearer and more idiomatic than `map` and `filter`

- Preserve the order of the original sequence

- Can also be lazy, returning an iterator rather than a new collection

````python
fruits = [("Kiwis", 40), ("Apples", 200), ("Pears", 150), ("Bananas", 120), ("Oranges", 90)]
double_lows = [2*c for p, c in fruits if c < 100]
print(double_lows)
````

````python
fruits = [("Kiwis", 40), ("Apples", 200), ("Pears", 150), ("Bananas", 120), ("Oranges", 90)]
double_lows = (2*c for p, c in fruits if c < 100) # lazy
print(double_lows)
print(tuple(double_lows))
````

````python
dict_low_fruits = {p : c for p, c in fruits if c < 100}
print(dict_low_fruits)
````

## ðŸ”¬ Closures

- A *closure* is a function object that remembers values from its enclosing scope, even if that scope has finished executing

- Typically arise when a function returns another function or is used as a callback

- Enclosed variables are captured by reference, not by value â€” the inner function keeps access to them

````python
def make_adder(n):
    def adder(x):
        return x + n  # remembers 'n'
    return adder

add5 = make_adder(5)
print(add5(10))
````

âš ï¸ Closures keep references to variables, not their values, which may be confusing

````python
funcs = []
# variable "i" changes in the loop
for i in range(3):
    def f():
        return i
    funcs.append(f)

print([fn() for fn in funcs])
````

````python
funcs = []
# to fix it, create a new variable "i" that has value "i"
for i in range(3):
    def f(i=i):
        return i
    funcs.append(f)

print([fn() for fn in funcs])
````

# Exception handling

- Exceptions signal *runtime* errors or exceptional conditions during execution

- Exceptions may be raised by Python automatically, or explicitly with `raise`

- Exceptions are handled using `try`â€“`except` blocks
  - Optionally `else` (executed if no exception occurs) and `finally` (always executed)

- Exceptions have types (e.g. `ValueError`, `TypeError`, `KeyError`) and can be caught selectively

- Unhandled exceptions propagate up the call stack and terminate the program

````python
def pos_int(s):
  n = int(s)
  if n < 0:
    raise ValueError("Not a positive number")
  return n
````

````python
# no exception
pos_int("10")
````

````python
# exception raised by int() propagated
pos_int("a10")
````

````python
# exception raised by pos_int()
pos_int("-10")
````

````python
# exception caught and printed gracefully
try:
  pos_int("-10")
except ValueError as e:
  print("Got an error! Message:", e)
````

# Basics of object-oriented programming in Python

- Classes define new types using the `class` keyword and group data with behavior

- Methods are functions defined inside a `class`; the first parameter is object on which the method was called, conventionally `self`

- By default, all methods are instance methods

- Object creation and initialization are handled by the constructor method `__init__`

- Python has no enforced visibility: naming conventions indicate intent
  - public (`x`), "protected" (`_x`), "name-mangled" to avoid clashes (`__x`)


````python
class Hero:
    # class-level attribute
    _heroes_created = 0

    def __init__(self, name, power):
        # instance attributes
        self.name = name
        self.__power = power
        Hero._heroes_created += 1

    # instance method
    def reveal_power(self):
        return f"{self.name} has power {self.__power}"

    # class method
    def total_heroes():
        return Hero._heroes_created

h1 = Hero("Arya", 95)
h2 = Hero("Jon", 90)

print(h1.reveal_power())          # instance method
print(Hero.total_heroes())        # class method
````

## Inheritance in Python

- A class can inherit from one or more parent classes, gaining their attributes and methods

- Child classes can override parent methods

- `super()` calls the parent method from the child

````python
# child class
class SuperHero(Hero):
    def reveal_power(self):
        base = super().reveal_power()  # call parent method
        return f"{base} and is unstoppable!"

h1 = SuperHero("Arya", 95)
print(h1.reveal_power())
````
